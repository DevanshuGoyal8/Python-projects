```python
import random
def playername():
    global first_player
    global second_player
    first_player = input("Please enter First Player Name : ")
    second_player = input("Please enter Second Player Name: ")
    return ([first_player,second_player])
   


```

```python
def start():
    global first_player
    global second_player
    global first_p_choice
    global X_set
    global O_set
    global final_dict
    X_set = []
    O_set = []
    first_p_sign = ''
    second_p_sign = ''
    final_dict = {}
    names = playername()
    toss()
    randomnum = random.randrange(1,10)
    fin_num = randomnum%2
    if_first_won = 0
    if fin_num == first_p_choice:
        print(f"Congrats {names[0]}, you have won the toss. Please choose your sign")
        if_first_won = 1
        first_p_sign = input("Choose between 'X' & 'O'")
        flag = 0
        while flag != 1:
            if first_p_sign == 'X':
                second_p_sign = 'O'
                flag = 1
            elif first_p_sign == 'O':
                second_p_sign = 'X'
                flag = 1
            else:
                flag = 0
    else:
        print(f"Congrats {names[1]}, you have won the toss. Please choose your sign")
        second_p_sign = input("Choose between 'X' & 'O'")
        flag = 0
        while flag != 1:

            if second_p_sign == 'X':
                first_p_sign = 'O'
                flag = 1
            elif second_p_sign == 'O':
                first_p_sign = 'X'
                flag = 1
            else:
                print("Something Wrong")
                flag = 0

    if if_first_won:
        print(f"{names[0]} has choosen {first_p_sign} And {names[1]} has choosen {second_p_sign}. As {names[0]} has won the toss, so he will start first")
        player_won = 1
    else : 
        print(f"{names[0]} has choosen {first_p_sign} And {names[1]} has choosen {second_p_sign}. As {names[1]} has won the toss, so he will start first")
        player_won = 2

    d_new = {}
    draw_board(d_new)
    game_over = 1
    while game_over!=0:

        option = int(input("Please choose the number where you want to put sign"))
        if player_won==1:
            d_new = {option: first_p_sign}
            player_won = 2
        else :
            d_new = {option: second_p_sign}
            player_won = 1

        draw_board(d_new)

        keys_list = list(final_dict.keys())
        keys_list.sort()
        keys_string = ''
        for keys in keys_list:
            keys_string = keys_string + str(keys)
        winner_list =  check_winner()
        if keys_string=='123456789':
            print ("Nobody Wins\n")
            play_again = input("Do you want to play again? Y or N")
            if play_again == 'Y':
                start()
            else :
                return
                
            game_over = 0
        elif  bool(winner_list):
            winner_sign = next(iter(winner_list))
            if winner_sign == first_p_sign:
                print (f"Congratulations {names[0]}, you have won")
            elif winner_sign == second_p_sign:
                print (f"Congratulations {names[1]}, you have won")
            else :
                print ("Nobody Wins\n")
               
            play_again = input("Do you want to play again? Y or N")
            if play_again == 'Y':
                start()
            else :
                return 
            game_over = 0
    
```

```python
def toss():
    global first_p_choice
    first_p_choice = int(input("Toss Time. Hey " + first_player + " Please enter 1 for odd and 0 for even: "))
    return int(first_p_choice)
```

```python
def draw_board(d):
    horizontal = '_'
    vertical = '|'
    it1 = [1,2,3]
    it2 = [1,2,3]
    return_string = ''
    global final_dict
    global X_set
    global Y_set
   
    final_dict.update(d)
    if not bool(final_dict):
        inc = 1
        for x in it1:
            row_string = ''
            if x!=3:
                for y in it2:
                    if y != 3:
                        row_string = row_string + horizontal + str(inc) + horizontal +  vertical
                    else :
                        row_string = row_string + horizontal + str(inc) + horizontal
                    inc = inc + 1
            else:
                for y in it2:
                    if y !=3 :
                        row_string = row_string + ' ' + str(inc) + ' ' + vertical
                    else:
                        row_string = row_string + ' ' + str(inc)
                    inc = inc + 1
            return_string = return_string + row_string + "\n"
    else:
        if (next(iter(d.values()))) == 'X':
            X_set.append(int(next(iter(d))))
        elif (next(iter(d.values()))) == 'O':
            O_set.append(int(next(iter(d))))
        inc = 1
        for x in it1:
            row_string = ''
            if x!=3:
                for y in it2:
                    if inc in final_dict.keys():
                        display = final_dict[inc]
                    else :
                        display = inc
                    if y != 3:
                        row_string = row_string + horizontal + str(display) + horizontal +  vertical
                    else :
                        row_string = row_string + horizontal + str(display) + horizontal
                    inc = inc + 1
            else:
                for y in it2:
                    if inc in final_dict.keys():
                        display = final_dict[inc]
                    else :
                        display = inc
                    if y !=3 :
                        row_string = row_string + ' ' + str(display) + ' ' + vertical
                    else:
                        row_string = row_string + ' ' + str(display)
                    inc = inc + 1
            return_string = return_string + row_string + "\n"
    print(return_string)
        
```

```python
def check_winner():
    winner_dict = {0:[1,2,3] ,1:[4,5,6], 2:[7,8,9], 3: [1,4,7], 4:[2,5,8], 5:[3,6,9],6:[1,5,9],7:[3,5,7]}
    global X_set
    global O_set
    for y in winner_dict.values():
        if(set(y).issubset(set(X_set))):
            return {'X':y}
        elif(set(y).issubset(set(O_set))):
            return {'O':y }
    return {}
```
